# **Dataset Transactions**

## **Revisión inicial del dataset**



```python
file_paths = ['./dataset/Transactions/102_Tran.csv', './dataset/Transactions/110_Tran.csv', './dataset/Transactions/103_Tran.csv', './dataset/Transactions/107_Tran.csv']
dfs = []

for file_path in file_paths:
    df = pd.read_csv(file_path, sep='|', header=None, names=["Date", "Store_id", "Customer_id", "Products"])
    dfs.append(df)

df_transactions = pd.concat(dfs, ignore_index=True)

# Convertir la columna 'Date' a tipo datetime
df_transactions['Date'] = pd.to_datetime(df_transactions['Date'], format='%Y-%m-%d')
```

### Análisis Detallado de Ventas: Diarias, Semanales y Mensuales

```python
# Preparar columnas de tiempo adicionales
df_transactions['Year'] = df_transactions['Date'].dt.year
df_transactions['Month'] = df_transactions['Date'].dt.month
df_transactions['Month_Name'] = df_transactions['Date'].dt.month_name()
df_transactions['Week'] = df_transactions['Date'].dt.isocalendar().week
df_transactions['DayOfWeek'] = df_transactions['Date'].dt.dayofweek  # 0=Lunes, 6=Domingo
df_transactions['DayName'] = df_transactions['Date'].dt.day_name()

print("Columnas de tiempo agregadas:")
print(df_transactions[['Date', 'Year', 'Month', 'Month_Name', 'Week', 'DayOfWeek', 'DayName']].head())
```

```python
# VENTAS DIARIAS
daily_sales = df_transactions.groupby('Date').agg({
    'Customer_id': 'count',      # Número de transacciones
    'Num_Products': 'sum',       # Total de productos vendidos
}).rename(columns={
    'Customer_id': 'Transacciones',
    'Num_Products': 'Productos_Vendidos',
})

print("="*60)
print("ANÁLISIS DE VENTAS DIARIAS")
print("="*60)
print(f"\nEstadísticas de ventas diarias:")
print(daily_sales.describe())
print(f"\nDía con más transacciones: {daily_sales['Transacciones'].idxmax().strftime('%Y-%m-%d')} ({daily_sales['Transacciones'].max():,} transacciones)")
print(f"Día con menos transacciones: {daily_sales['Transacciones'].idxmin().strftime('%Y-%m-%d')} ({daily_sales['Transacciones'].min():,} transacciones)")
print(f"\nPromedio diario de transacciones: {daily_sales['Transacciones'].mean():.0f}")
print(f"Promedio diario de productos vendidos: {daily_sales['Productos_Vendidos'].mean():.0f}")

print("\n" + "="*60)
print("Top 10 días con más ventas:")
print("="*60)
display(daily_sales.sort_values('Transacciones', ascending=False).head(10))
```

```python
# VENTAS SEMANALES
weekly_sales = df_transactions.groupby(['Year', 'Week']).agg({
    'Customer_id': 'count',
    'Num_Products': 'sum'
}).rename(columns={
    'Customer_id': 'Transacciones',
    'Num_Products': 'Productos_Vendidos'
})

print("="*60)
print("ANÁLISIS DE VENTAS SEMANALES")
print("="*60)
print(f"\nEstadísticas de ventas semanales:")
print(weekly_sales.describe())
print(f"\nPromedio semanal de transacciones: {weekly_sales['Transacciones'].mean():.0f}")
print(f"Promedio semanal de productos vendidos: {weekly_sales['Productos_Vendidos'].mean():.0f}")

print("\n" + "="*60)
print("Top 10 semanas con más ventas:")
print("="*60)
display(weekly_sales.sort_values('Transacciones', ascending=False).head(10))
```

```python
# VENTAS MENSUALES
monthly_sales = df_transactions.groupby(['Year', 'Month', 'Month_Name']).agg({
    'Customer_id': 'count',
    'Num_Products': 'sum'
}).rename(columns={
    'Customer_id': 'Transacciones',
    'Num_Products': 'Productos_Vendidos'
})

print("="*60)
print("ANÁLISIS DE VENTAS MENSUALES")
print("="*60)
print(f"\nEstadísticas de ventas mensuales:")
print(monthly_sales.describe())
print(f"\nPromedio mensual de transacciones: {monthly_sales['Transacciones'].mean():.0f}")
print(f"Promedio mensual de productos vendidos: {monthly_sales['Productos_Vendidos'].mean():.0f}")

print("\n" + "="*60)
print("Ventas por mes (ordenado por volumen):")
print("="*60)
display(monthly_sales.sort_values('Transacciones', ascending=False))
```

```python
# Visualización de ventas diarias, semanales y mensuales
fig, axes = plt.subplots(3, 1, figsize=(16, 12))

# Ventas diarias
axes[0].plot(daily_sales.index, daily_sales['Transacciones'], color='blue', linewidth=1.5, alpha=0.7)
axes[0].fill_between(daily_sales.index, daily_sales['Transacciones'], alpha=0.3, color='blue')
axes[0].set_xlabel('Fecha', fontsize=12)
axes[0].set_ylabel('Número de Transacciones', fontsize=12)
axes[0].set_title('Evolución de Ventas Diarias', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].axhline(y=daily_sales['Transacciones'].mean(), color='red', linestyle='--', 
                label=f'Promedio: {daily_sales["Transacciones"].mean():.0f}', linewidth=2)
axes[0].legend()

# Ventas semanales
weekly_sales_reset = weekly_sales.reset_index()
weekly_sales_reset['Week_Label'] = weekly_sales_reset['Year'].astype(str) + '-W' + weekly_sales_reset['Week'].astype(str)
axes[1].bar(range(len(weekly_sales_reset)), weekly_sales_reset['Transacciones'], 
            color='green', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Semana', fontsize=12)
axes[1].set_ylabel('Número de Transacciones', fontsize=12)
axes[1].set_title('Evolución de Ventas Semanales', fontsize=14, fontweight='bold')
axes[1].axhline(y=weekly_sales['Transacciones'].mean(), color='red', linestyle='--', 
                label=f'Promedio: {weekly_sales["Transacciones"].mean():.0f}', linewidth=2)
axes[1].grid(True, alpha=0.3, axis='y')
axes[1].legend()

# Ventas mensuales
monthly_sales_reset = monthly_sales.reset_index()
axes[2].bar(range(len(monthly_sales_reset)), monthly_sales_reset['Transacciones'], 
            color='orange', edgecolor='black', alpha=0.7)
axes[2].set_xticks(range(len(monthly_sales_reset)))
axes[2].set_xticklabels(monthly_sales_reset['Month_Name'], rotation=45, ha='right')
axes[2].set_xlabel('Mes', fontsize=12)
axes[2].set_ylabel('Número de Transacciones', fontsize=12)
axes[2].set_title('Evolución de Ventas Mensuales', fontsize=14, fontweight='bold')
axes[2].axhline(y=monthly_sales['Transacciones'].mean(), color='red', linestyle='--', 
                label=f'Promedio: {monthly_sales["Transacciones"].mean():.0f}', linewidth=2)
axes[2].grid(True, alpha=0.3, axis='y')
axes[2].legend()

plt.tight_layout()
plt.show()
```

### 3. Segmentación de clientes (RFM Analysis)

```python
# Análisis RFM (Recency, Frequency, Monetary)
# Recency: Días desde la última compra
# Frequency: Número de compras
# Monetary: En este caso, usaremos el número de productos comprados

fecha_referencia = df_transactions['Date'].max() + pd.Timedelta(days=1)

rfm = df_transactions.groupby('Customer_id').agg({
    'Date': [lambda x: (fecha_referencia - x.max()).days, 'count'],  # Recency y Frequency
    'Products': lambda x: x.str.split().apply(len).sum()  # Monetary (total productos - separados por espacios)
})

# Aplanar las columnas
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm.reset_index()

print("Primeros registros del análisis RFM:")
print(rfm.head(10))
print("\nEstadísticas RFM:")
print(rfm[['Recency', 'Frequency', 'Monetary']].describe())
```

**¿Qué es el Análisis RFM?**

El **RFM Analysis** es una técnica de segmentación de clientes que evalúa tres dimensiones clave del comportamiento de compra:

#### **Las 3 Métricas RFM:**

1. **R - Recency (Recencia)**: ¿Cuán recientemente compró el cliente?
   - Medida en **días desde la última compra**
   - **Menor es mejor**: Un cliente que compró ayer es más valioso que uno que compró hace meses
   - Indica qué tan "activo" está el cliente en este momento

2. **F - Frequency (Frecuencia)**: ¿Con qué frecuencia compra?
   - Medida en **número total de transacciones**
   - **Mayor es mejor**: Más compras = mayor lealtad
   - Indica el nivel de compromiso con la marca

3. **M - Monetary (Monetario)**: ¿Cuánto gasta?
   - Medida en **valor total gastado** (en este caso, número de productos comprados)
   - **Mayor es mejor**: Clientes que gastan más son más valiosos
   - Indica el valor económico del cliente

#### **¿Cómo funciona?**

**Paso 1**: Se calculan las 3 métricas para cada cliente

**Paso 2**: Se asigna un **score de 1 a 5** a cada métrica (dividiendo clientes en quintiles):
- **Score 5**: Los mejores (top 20%)
- **Score 1**: Los peores (bottom 20%)
- *Nota*: Para Recency se invierte (menor días = score 5)

**Paso 3**: Se crea un **score combinado** (ej: "543" = R:5, F:4, M:3) y se suma (5+4+3=12)

**Paso 4**: Se **segmenta** según el score total:
- **Champions (13-15)**: Los mejores clientes → Recompensar, programas VIP
- **Loyal Customers (10-12)**: Clientes leales → Mantener enganchados
- **Potential Loyalists (7-9)**: Con potencial → Incentivos, membresías
- **At Risk (5-6)**: En riesgo de abandonar → Campañas urgentes de reactivación
- **Lost (3-4)**: Clientes perdidos → Intentar reconquista o dejar ir

```python
# Crear scores RFM (1-5, donde 5 es mejor)
# Para Recency, menor es mejor, así que invertimos
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])

# Crear score RFM combinado
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score_Sum'] = rfm[['R_Score', 'F_Score', 'M_Score']].astype(int).sum(axis=1)

print("RFM con scores:")
rfm.head(10)
```

```python
# Segmentación basada en RFM Score
def segmentar_cliente(row):
    if row['RFM_Score_Sum'] >= 13:
        return 'Champions'
    elif row['RFM_Score_Sum'] >= 10:
        return 'Loyal Customers'
    elif row['RFM_Score_Sum'] >= 7:
        return 'Potential Loyalists'
    elif row['RFM_Score_Sum'] >= 5:
        return 'At Risk'
    else:
        return 'Lost'

rfm['Segmento'] = rfm.apply(segmentar_cliente, axis=1)

# Contar clientes por segmento
segmento_counts = rfm['Segmento'].value_counts()
print("Distribución de clientes por segmento:")
print(segmento_counts)
print(f"\nPorcentaje por segmento:")
print((segmento_counts / segmento_counts.sum() * 100).round(2))
```

```python
# Visualización de segmentación
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gráfico de barras de segmentos
axes[0, 0].bar(segmento_counts.index, segmento_counts.values, color='steelblue', edgecolor='black')
axes[0, 0].set_xlabel('Segmento')
axes[0, 0].set_ylabel('Número de Clientes')
axes[0, 0].set_title('Distribución de Clientes por Segmento')
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, alpha=0.3)

# Scatter plot: Recency vs Frequency
scatter = axes[0, 1].scatter(rfm['Recency'], rfm['Frequency'], 
                            c=rfm['RFM_Score_Sum'], cmap='RdYlGn', 
                            alpha=0.6, edgecolors='black', linewidth=0.5)
axes[0, 1].set_xlabel('Recency (días)')
axes[0, 1].set_ylabel('Frequency (compras)')
axes[0, 1].set_title('Recency vs Frequency')
plt.colorbar(scatter, ax=axes[0, 1], label='RFM Score')
axes[0, 1].grid(True, alpha=0.3)

# Scatter plot: Frequency vs Monetary
scatter2 = axes[1, 0].scatter(rfm['Frequency'], rfm['Monetary'], 
                             c=rfm['RFM_Score_Sum'], cmap='RdYlGn', 
                             alpha=0.6, edgecolors='black', linewidth=0.5)
axes[1, 0].set_xlabel('Frequency (compras)')
axes[1, 0].set_ylabel('Monetary (productos)')
axes[1, 0].set_title('Frequency vs Monetary')
plt.colorbar(scatter2, ax=axes[1, 0], label='RFM Score')
axes[1, 0].grid(True, alpha=0.3)

# Box plot de RFM Score por segmento
rfm.boxplot(column='RFM_Score_Sum', by='Segmento', ax=axes[1, 1])
axes[1, 1].set_xlabel('Segmento')
axes[1, 1].set_ylabel('RFM Score Total')
axes[1, 1].set_title('RFM Score por Segmento')
axes[1, 1].tick_params(axis='x', rotation=45)
plt.suptitle('')

plt.tight_layout()
plt.show()
```

### 2. Análisis de Reglas de Asociación (Market Basket Analysis)

Preparamos los datos para el análisis de Market Basket. Convertimos cada transacción (que contiene productos separados por espacios) en una lista de productos individuales. Esto nos permite analizar qué productos se compran juntos con frecuencia. 

El resultado muestra que tenemos 1,108,987 transacciones con 449 productos únicos.

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Preparar datos en formato de transacciones (lista de listas)
# Los productos están separados por ESPACIOS
transactions = []
for products_str in df_transactions['Products']:
    # Dividir por espacios
    transaction = products_str.split()
    if transaction:  # Solo agregar si tiene productos válidos
        transactions.append(transaction)

print(f"Total de transacciones: {len(transactions)}")
print(f"Ejemplo de transacción: {transactions[0]}")

# Calcular productos únicos
unique_products = set()
for t in transactions:
    unique_products.update(t)
print(f"Productos únicos en transacciones: {len(unique_products)}")
```

```python
# Codificar transacciones en formato one-hot
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print(f"Dimensiones del dataset codificado: {df_encoded.shape}")
print("\nPrimeras filas del dataset codificado:")
df_encoded.head()
```

**Explicación**: Aquí transformamos las transacciones a formato **one-hot encoding** (matriz binaria). Cada fila representa una transacción y cada columna un producto, con valores True/False indicando si el producto está en esa transacción. Este formato es requerido por el algoritmo Apriori para identificar patrones de compra.

**Nota sobre el Algoritmo Apriori**

El algoritmo Apriori busca patrones frecuentes de productos que se compran juntos. Debido al **gran tamaño del dataset** (1,108,987 transacciones), utilizamos las siguientes optimizaciones:

- **`min_support=0.02` (2%)**: Solo consideramos combinaciones que aparecen en al menos 2% de las transacciones (~22,180 transacciones). Esto reduce significativamente el número de combinaciones a analizar.
- **`low_memory=True`**: Optimiza el algoritmo para usar menos memoria RAM.
- **`verbose=1`**: Muestra el progreso del algoritmo.

Un soporte de 2% es razonable para este dataset y nos permitirá identificar las asociaciones más significativas sin saturar la memoria.

```python
# Encontrar itemsets frecuentes usando Apriori
# min_support: soporte mínimo (aumentado para reducir uso de memoria)
# low_memory=True: optimiza el uso de memoria
# NOTA: Con dataset grande (1M+ transacciones), necesitamos un soporte más alto

print("Ejecutando algoritmo Apriori (esto puede tomar algunos minutos)...")
frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True, low_memory=True, verbose=1)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

print(f"\nTotal de itemsets frecuentes encontrados: {len(frequent_itemsets)}")
print("\nTop 20 itemsets más frecuentes:")
frequent_itemsets.sort_values('support', ascending=False).head(20)
```

**Explicación**: El algoritmo **Apriori** identifica conjuntos de productos que aparecen frecuentemente juntos. Con `min_support=0.02`, buscamos combinaciones que aparecen en al menos 2% de las transacciones (~22,180 veces). Los resultados muestran tanto productos individuales frecuentes como combinaciones de 2 o más productos que los clientes tienden a comprar juntos.

```python
# Generar reglas de asociación
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules['antecedent_len'] = rules['antecedents'].apply(lambda x: len(x))
rules['consequent_len'] = rules['consequents'].apply(lambda x: len(x))

print(f"Total de reglas de asociación encontradas: {len(rules)}")
print("\nTop 20 reglas por confianza:")
rules.sort_values('confidence', ascending=False).head(20)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

**Explicación**: Generamos **reglas de asociación** en formato "Si compra A → entonces compra B". La métrica `confidence >= 0.3` significa que buscamos reglas donde al menos el 30% de las veces que se compra el producto A, también se compra el producto B. Estas reglas son útiles para recomendaciones de productos, promociones cruzadas y diseño de tienda.

```python
# Filtrar reglas por lift > 1 (indica asociación positiva)
strong_rules = rules[rules['lift'] > 1].sort_values('lift', ascending=False)

print(f"Reglas con lift > 1: {len(strong_rules)}")
print("\nTop 20 reglas por lift:")
strong_rules.head(20)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

**Explicación**: Filtramos las reglas por **Lift > 1**, que indica **asociación positiva real**. Un Lift > 1 significa que los productos se compran juntos con más frecuencia de lo que sería esperado por azar. Estas son las reglas más valiosas porque representan patrones genuinos de comportamiento de compra, no coincidencias aleatorias.

```python
# Visualización de reglas de asociación
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Scatter plot: Support vs Confidence
scatter1 = axes[0, 0].scatter(rules['support'], rules['confidence'], 
                              c=rules['lift'], cmap='viridis', 
                              alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
axes[0, 0].set_xlabel('Support')
axes[0, 0].set_ylabel('Confidence')
axes[0, 0].set_title('Support vs Confidence (coloreado por Lift)')
plt.colorbar(scatter1, ax=axes[0, 0], label='Lift')
axes[0, 0].grid(True, alpha=0.3)

# Scatter plot: Support vs Lift
scatter2 = axes[0, 1].scatter(rules['support'], rules['lift'], 
                              c=rules['confidence'], cmap='plasma', 
                              alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
axes[0, 1].set_xlabel('Support')
axes[0, 1].set_ylabel('Lift')
axes[0, 1].set_title('Support vs Lift (coloreado por Confidence)')
plt.colorbar(scatter2, ax=axes[0, 1], label='Confidence')
axes[0, 1].grid(True, alpha=0.3)

# Scatter plot: Confidence vs Lift
scatter3 = axes[1, 0].scatter(rules['confidence'], rules['lift'], 
                              c=rules['support'], cmap='coolwarm', 
                              alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
axes[1, 0].set_xlabel('Confidence')
axes[1, 0].set_ylabel('Lift')
axes[1, 0].set_title('Confidence vs Lift (coloreado por Support)')
plt.colorbar(scatter3, ax=axes[1, 0], label='Support')
axes[1, 0].grid(True, alpha=0.3)

# Histograma de Lift
axes[1, 1].hist(rules['lift'], bins=50, edgecolor='black', color='orange')
axes[1, 1].axvline(x=1, color='red', linestyle='--', linewidth=2, label='Lift = 1')
axes[1, 1].set_xlabel('Lift')
axes[1, 1].set_ylabel('Número de Reglas')
axes[1, 1].set_title('Distribución de Lift')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

**Explicación**: Esta visualización explora las relaciones entre las tres métricas clave:
- **Support vs Confidence**: Muestra qué tan frecuentes son las combinaciones y qué tan confiables son las reglas
- **Support vs Lift**: Identifica reglas frecuentes con alta asociación positiva
- **Confidence vs Lift**: Diferencia entre reglas confiables y reglas con verdadera asociación
- **Distribución de Lift**: La línea roja en Lift=1 separa asociaciones positivas (derecha) de negativas/neutrales (izquierda)

**Interpretación de las métricas:**

- **Support (Soporte)**: Proporción de transacciones que contienen el itemset. Indica qué tan frecuente es la combinación.
- **Confidence (Confianza)**: Probabilidad de que se compre el consecuente dado que se compró el antecedente.
- **Lift**: Mide cuánto más probable es comprar el consecuente cuando se compra el antecedente, comparado con comprarlo al azar.
  - Lift > 1: Indica asociación positiva (los productos se compran juntos más de lo esperado)
  - Lift = 1: No hay asociación
  - Lift < 1: Asociación negativa (comprar uno reduce la probabilidad de comprar el otro)
